```python
import numpy as np
```

```python
graph = []
min_val = 100

for i in range(1,10000000):
    x = np.random.rand()
    y = np.random.rand()
    val1 = (x+y)/(x*1.0)
    val2 = x/(y*1.0)
    diff = abs(val1 - val2)
    if(diff<min_val):
        min_val = diff
        graph.append(val2)
        gr = val2
```

```python
plt.plot(graph)
plt.show()
```
